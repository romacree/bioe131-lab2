### Research and write-up

Antibiotic resistant bacteria has become a problem that is rapidly growing, particularly multidrug resistant bacteria (resistant to more than two drugs). From 1997-2002 the percentage of multidrug-resistant strains increased from 13-22%. Pseudomonas is a genus of bacteria that is able to rapidly mutate through the time frame of treatment for antibiotics. It is extremely resilient and survives in areas such as swimming pools, contact lens solution, humidifiers, soil and vegetables. It is also thrives in hospitals, including respiratory equipment, antiseptics, soap, sinks, mops, medicine, and hydrotherapy pools.

According to our KEGG map for our sequence when using the RAST, 100% of Pseudomonas species are beta-Lactam resistant, which includes resistance to antibiotics such as penicillin. There are multiple pathways for beta-Lactam resistance to be found, including pathways caused by Cephamycin C and Cephalosporin C. There are 14 distinct ECs that control the biosynthesis of penicillin and cephalosporin, 4 of which are found in Pseudomonas. We also observed 4 distinct ECs for biosynthesis of Tetracycline, indicating the possibilities of resistance to Tetracycline antibiotics. This is consistent with a paper relating the functionality of the efflux pump to be causal for antibiotic resistance. The efflux pump is a transmembrane protein that allows amphiphilic drugs into the cell, which when mutated, nullifies a variety of antibiotics such as fluoroquinolones, tetracycline, chloramphenicol, macrolides.

The genome of Pseudomonas has multiple antibiotic resistant genes that influence multiple pathways, making effective treatment for complex. Treatment for this bacteria can be complex because the rapid rate of mutation can easily exceed the time frame for antibiotic treatment for the patient. Pseudomonas also is able to thrive and is more likely to become resistant in immunocompromised patients, making this bacteria extremely dangerous. The strategy for treatment is generally to give multiple types of antibiotics, however this poses a pressing concern for evolutionary strains that are multidrug resistant.

Source: Lister, Philip D., et al. “Antibacterial-Resistant Pseudomonas Aeruginosa: Clinical Impact and Complex Regulation of Chromosomally Encoded Resistance Mechanisms.” Clinical Microbiology Reviews, American Society for Microbiology (ASM), Oct. 2009, www.ncbi.nlm.nih.gov/pmc/articles/PMC2772362/#r234.
